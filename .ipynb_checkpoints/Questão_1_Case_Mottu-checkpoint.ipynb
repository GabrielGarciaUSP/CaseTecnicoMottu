```python
import pandas as pd

# Coloca a base de dados em uma variável
arquivo = "base_dados.xlsx"

# Lê o arquivo para ver os nomes das abas
xls = pd.ExcelFile(arquivo)

print("--- AS ABAS DESTE ARQUIVO SÃO: ---")
print(xls.sheet_names)
```

```python
# Carregamos a aba com os dados completos dos funcionários
df = pd.read_excel("base_dados.xlsx", sheet_name='Q1_turnover_employees_FULL')

print("--- COLUNAS DA ABA FULL ---")
print(df.columns.tolist())

print("\n--- AMOSTRA DOS DADOS ---")
display(df.head(3))
```

```python
# Arrumamos as datas antes de tudo, para deixar nossa base de dados preparada
df['data_admissao'] = pd.to_datetime(df['data_admissao'], errors='coerce')
df['data_demissao'] = pd.to_datetime(df['data_demissao'], errors='coerce')
df['data_certificacao'] = pd.to_datetime(df['data_certificacao'], errors='coerce')
```

```python
# Definimos um filtro, para buscar apenas os mecânicos
df_mec = df[df['cargo'] == 'mecanico'].copy()
```

```python
# Definimos a linha do tempo, sendo de Jan/24 a Jun/25
meses_analise = pd.date_range(start='2024-01-01', end='2025-06-01', freq='MS')
resultados = []
```

```python
for mes in meses_analise:
    # Definimos o fim do mês
    fim_mes = mes + pd.offsets.MonthEnd(0)
    
    # Verificação de quem entrou antes do fim do mês e (não saiu ou saiu depois do dia 1 desse mês)
    ativos_no_mes = df_mec[
        (df_mec['data_admissao'] <= fim_mes) &
        ((df_mec['data_demissao'].isna()) | (df_mec['data_demissao'] >= mes))
    ]
    
    # Verificação de quem saiu no mesmo mês
    saidas_no_mes = df_mec[
        (df_mec['data_demissao'] >= mes) &
        (df_mec['data_demissao'] <= fim_mes)
    ]
    
    # Separação dos grupos de mecânicos se baseando na Data de Certificação
    
    # Faixa Preta vs Normais
    # É Faixa Preta se tem data de certificação e ela aconteceu antes desse mês acabar
    ativos_fp = ativos_no_mes[ativos_no_mes['data_certificacao'] <= fim_mes]
    ativos_normal = ativos_no_mes[~(ativos_no_mes['data_certificacao'] <= fim_mes)] # O resto
    
    # Definir saidas de pessoas normais e de mecânicos faixa preta
    saidas_fp = saidas_no_mes[saidas_no_mes['data_certificacao'] <= fim_mes]
    saidas_normal = saidas_no_mes[~(saidas_no_mes['data_certificacao'] <= fim_mes)]
    
    # Turnover geral dos mecânicos
    turnover_geral = len(saidas_no_mes) / len(ativos_no_mes) if len(ativos_no_mes) > 0 else 0
    
    # Turnover Faixa Preta
    turnover_fp = len(saidas_fp) / len(ativos_fp) if len(ativos_fp) > 0 else 0
    
    # Turnover Normal
    turnover_normal = len(saidas_normal) / len(ativos_normal) if len(ativos_normal) > 0 else 0
    
    resultados.append({
        'Mes': mes.strftime('%Y-%m'), # Formato Bonito (2024-01)
        'Ativos_Total': len(ativos_no_mes),
        'Turnover_Geral': turnover_geral,
        'Turnover_FaixaPreta': turnover_fp,
        'Turnover_Normal': turnover_normal,
        'Qtd_Saidas_Normal': len(saidas_normal),
        'Qtd_Saidas_FP': len(saidas_fp)
    })

# Criação da tabela final
df_final = pd.DataFrame(resultados)

# Printar a tabela
display(df_final)

# Importação da bibliote
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_final['Mes'], df_final['Turnover_Normal'], label='Mecânicos Normais', color='red', marker='o')
plt.plot(df_final['Mes'], df_final['Turnover_FaixaPreta'], label='Mecânicos Faixa Preta', color='green', marker='o')
plt.plot(df_final['Mes'], df_final['Turnover_Geral'], label='Média Geral', color='gray', linestyle='--')

plt.title('Comparativo de Turnover: Faixa Preta vs. Normais')
plt.ylabel('Taxa de Turnover')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
df_final = df_final.sort_values('Mes')

# Se tiver duplicatas exatas, removemos 
df_final = df_final.drop_duplicates(subset=['Mes'])

# Print do gráfico 
plt.figure(figsize=(12, 6))

plt.plot(df_final['Mes'], df_final['Turnover_Normal'], label='Mecânicos Normais', color='red', marker='o', linewidth=2)
plt.plot(df_final['Mes'], df_final['Turnover_FaixaPreta'], label='Mecânicos Faixa Preta', color='green', marker='o', linewidth=2)
plt.plot(df_final['Mes'], df_final['Turnover_Geral'], label='Média Geral', color='gray', linestyle='--', alpha=0.7)

plt.title('Turnover: Faixa Preta vs. Normais', fontsize=14)
plt.ylabel('Taxa de Turnover', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Percebe-se que teve um grande pico de turnover dos mecanicos faixa preta nos meses de Janeiro Fevereiro e março
# Logo vamos fazer uma análise para provar a hipótese 1 

print("--- Análise do pico de JAN a MAR 2025 ---")
print(f"{'MÊS':<10} | {'ATIVOS (Faixa Preta)':<20} | {'SAÍDAS (Faixa Preta)':<20} | {'TURNOVER CALCULADO'}")
print("-" * 75)

meses_investigacao = pd.date_range(start='2025-01-01', end='2025-04-01', freq='MS')

for mes in meses_investigacao:
    fim_mes = mes + pd.offsets.MonthEnd(0)
    
    # Filtra Faixa Pretas ativos naquele mês
    ativos_fp = df_mec[
        (df_mec['data_admissao'] <= fim_mes) &
        ((df_mec['data_demissao'].isna()) | (df_mec['data_demissao'] >= mes)) &
        (df_mec['data_certificacao'] <= fim_mes) # Só quem já tinha o certificado na época
    ]
    
    # Filtra Faixa Pretas que saíram naquele mês
    saidas_fp = df_mec[
        (df_mec['data_demissao'] >= mes) &
        (df_mec['data_demissao'] <= fim_mes) &
        (df_mec['data_certificacao'] <= fim_mes)
    ]
    
    qtd_ativos = len(ativos_fp)
    qtd_saidas = len(saidas_fp)
    turnover = (qtd_saidas / qtd_ativos) * 100 if qtd_ativos > 0 else 0
    
    print(f"{mes.strftime('%Y-%m'):<10} | {qtd_ativos:<20} | {qtd_saidas:<20} | {turnover:.2f}%")
```

```python
# Vamos classificar cada saída pelo tempo de casa que a pessoa tinha

# Filtramos apenas quem saiu (demitidos/pediram conta)
demissoes = df_mec[df_mec['data_demissao'].notna()].copy()

# Calculamos quantos meses a pessoa durou na empresa dividindo por 30 para ter o número aproximado de meses
demissoes['meses_duracao'] = (demissoes['data_demissao'] - demissoes['data_admissao']).dt.days / 30

# Criamos categorias para o gráfico, sendo 0-3 meses, 3-6 meses, 6-12 meses, 1 ano+
bins = [0, 3, 6, 12, 100]
labels = ['0-3 Meses (Risco Crítico)', '3-6 Meses', '6-12 Meses', '1 Ano+']
demissoes['faixa_tempo_casa'] = pd.cut(demissoes['meses_duracao'], bins=bins, labels=labels)

# Contamos quantas pessoas saíram em cada faixa
analise_safra = demissoes['faixa_tempo_casa'].value_counts().sort_index()

# Print do gráfico
plt.figure(figsize=(10, 6))
barras = plt.bar(analise_safra.index, analise_safra.values, color=['#d62728', '#ff7f0e', '#fdb462', '#2ca02c']) # Vermelho para crítico

# Adicionar os números em cima das barras
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2., altura,
             f'{int(altura)}',
             ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title('Onde está o sangramento? (Volume de Saídas por Tempo de Casa)', fontsize=14)
plt.ylabel('Quantidade de Pessoas que Saíram', fontsize=12)
plt.grid(axis='y', alpha=0.3)

plt.show()

print(analise_safra)
```

```python
# Analisando os motivos de saída, vamos ver o top 10 motivos
print("--- Principais motivos de saída ---")
motivos = df_mec['motivo_saida'].value_counts(dropna=False).head(10)
display(motivos)

# Vamos cruzar Motivo com Tempo de Casa, para responder quem sai com 6-12 meses, e qual o motivo?
print("\n--- MOTIVOS DE QUEM SAI ENTRE 6-12 MESES ---")

# Filtra só quem saiu nessa faixa critica
grupo_risco = df_mec[
    (df_mec['data_demissao'].notna()) &
    ((df_mec['data_demissao'] - df_mec['data_admissao']).dt.days > 180) & # Mais de 6 meses
    ((df_mec['data_demissao'] - df_mec['data_admissao']).dt.days <= 365)  # Menos de 1 ano
]

motivos_risco = grupo_risco['motivo_saida'].value_counts().head(5)
display(motivos_risco)
```

```python
# Analisando os motivos de saída, vamos ver os motivos
# 1. Filtramos o "Grupo de Risco" (Quem sai entre 6 e 12 meses)
grupo_risco = df_mec[
    (df_mec['data_demissao'].notna()) &
    ((df_mec['data_demissao'] - df_mec['data_admissao']).dt.days > 180) & # Mais de 6 meses
    ((df_mec['data_demissao'] - df_mec['data_admissao']).dt.days <= 365)  # Menos de 1 ano
]

# Contamos os motivos 
motivos_count = grupo_risco['motivo_saida'].value_counts().head(5).sort_values(ascending=True)

# Print do gráfico mas na horizontal
plt.figure(figsize=(10, 6))

barras = plt.barh(motivos_count.index, motivos_count.values, color='#ff7f0e')

for index, value in enumerate(motivos_count.values):
    plt.text(value, index, f' {value}', va='center', fontweight='bold', fontsize=11)

plt.title('Por que perdemos quem tem 6-12 meses de casa?', fontsize=14)
plt.xlabel('Quantidade de Saídas', fontsize=12)
plt.grid(axis='x', alpha=0.3)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
